```python
import pandas as pd

# specifing the filenames in the desired order
# loading the labelled datasets into one df

filenames = [
    "100_politifact_labelled_GPT4_nNVO.csv",
    "200_politifact_labelled_GPT4_nNVO.csv",
    "300_politifact_labelled_GPT4_nNVO.csv",
    "400_politifact_labelled_GPT4_nNVO.csv",
    "420_politifact_labelled_GPT4_nNVO.csv",
    "500_politifact_labelled_GPT4_nNVO.csv",
    "600_politifact_labelled_GPT4_nNVO.csv",
    "700_politifact_labelled_GPT4_nNVO1.csv",
    "800_politifact_labelled_GPT4_nNVO.csv",
    "900_politifact_labelled_GPT4_nNVO.csv",
    "1001_politifact_labelled_GPT4_nNVO.csv",
]

# Initialize an empty list to store DataFrames
dfs = []

# Iterate over each filename
for filename in filenames:
    # Read CSV file into a DataFrame
    df = pd.read_csv(filename)
    
    # Append DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames in the list along the rows axis
combined_df = pd.concat(dfs, axis=0, ignore_index=True)

# Display the combined DataFrame
display(combined_df)
```

Checking how many rows are missing in the labelling

```python

# Check if there are non-null values in the "Fact assessment" column
mask = combined_df['Fact assessment'].notnull()

# Create a subset of the DataFrame where "Fact assessment" column has non-null values
subset_df = combined_df[mask]

# Drop duplicate rows based on values in the "Unnamed: 0" column within the subset

# ONLY KEEPING THE ONE THAT ARE NOT UNIQUE
subset_df.drop_duplicates(subset=['Unnamed: 0'], inplace=True)

# Display the DataFrame after dropping duplicates
display(subset_df)
```

```python
# Rename columns in rearranged_df
subset_df.rename(columns={'Toxicity': 'Toxicity_GPT4_without_NV', 'Fact assessment': 'Fact assessment_GPT4_without_NV'}, inplace=True)

# Keep only the specified columns in rearranged_df
subset_df = subset_df[['statement','Unnamed: 0', 'Toxicity_GPT4_without_NV', 'Fact assessment_GPT4_without_NV']]

# Display the final rearranged DataFrame
display(subset_df)
```

Stacking with the missing rows which were labelled separately



```python
missing_df = pd.read_csv("Politifact_labelled_GPT4_without_NV_missing_rows.csv")
#display(missing_df)


missing_df = missing_df[['statement_x','Unnamed: 0','Toxicity', 'Fact assessment']]
missing_df.rename(columns={"statement_x": 'statement','Toxicity': 'Toxicity_GPT4_without_NV', 'Fact assessment': 'Fact assessment_GPT4_without_NV'}, inplace=True)
display(missing_df)
```

```python
# Concatenating the dataframes
df_combined = pd.concat([subset_df, missing_df], ignore_index=True)
display(df_combined)
subset_df=df_combined
```

EXTRACTING THE ORDER OF THE DATASET FROM A DATASET THAT WAS LABELLED WITH GPT3 


```python


# Read the file "1000_politifact_labelled_GPT3_wNV.csv" to extract the order of observations
order_df = pd.read_csv("1000_politifact_labelled_GPT3_wNV.csv")

# Rename columns
order_df.rename(columns={'Toxicity': 'Toxicity_GPT3_wNV', 'Fact assessment': 'Fact_assessment_GPT3_wNV'}, inplace=True)

# Display the modified order_df DataFrame
display(order_df)
```

```python
rearranged_df = pd.merge(order_df, subset_df[['statement','Unnamed: 0', 'Toxicity_GPT4_without_NV', 'Fact assessment_GPT4_without_NV']], 
                         how='left', on='Unnamed: 0')

# Display the final rearranged DataFrame
display(rearranged_df)
rearranged_df.to_csv("new_df.csv")
```
